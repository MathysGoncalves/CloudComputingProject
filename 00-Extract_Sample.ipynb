```python
import pandas as pd

import gc
pd.set_option('display.max_columns', None)
```

## Loading Data

On garde l'ensemble des données et on concat

```python
def load_df():
  
  df_2016 = pd.read_csv("full_2016.csv")
  df_2017 = pd.read_csv("full_2017.csv")
  df_2018 = pd.read_csv("full_2018.csv")
  df_2019 = pd.read_csv("full_2019.csv")
  df_2020 = pd.read_csv('full_2020.csv')
  print(len(df_2016.columns), len(df_2017.columns), len(df_2018.columns), len(df_2019.columns), len(df_2020.columns))

  df = pd.concat([df_2016, df_2017, df_2018, df_2019, df_2020], ignore_index=True)

  del df_2016
  del df_2017
  del df_2018
  del df_2019
  del df_2020
  gc.collect()

  return df
```

```python
df = load_df()
```

## Data Preparation

Filtre uniquement sur les vente.

```python
def modif(df):

  mask_nature = df['nature_mutation'] == 'Vente'
  df = df[mask_nature]
  
  df['date_mutation'] = pd.to_datetime(df['date_mutation'])
  df['Month'] = df['date_mutation'].dt.month
  df['Year'] = df['date_mutation'].dt.year

  df = df[['valeur_fonciere', 'code_departement', 'surface_reelle_bati', 'nombre_pieces_principales', 'surface_terrain', 'type_local', 'Month', 'Year', 'code_commune']]
  
  df['code_commune'] = df['code_commune'].astype(str)
  df['code_departement'] = df['code_departement'].astype(str)
  mask_dep = (df['code_departement'] == '18') | (df['code_departement'] == '28') | (df['code_departement'] == '36') | (df['code_departement'] == '37') | (df['code_departement'] == '41') | (df['code_departement'] == '45')
  df = df[mask_dep]

  df['surface_terrain'] = df['surface_terrain'].fillna(0)
  df = df.dropna()
  df = df.drop_duplicates()

  return df
```

```python
df_prep = modif(df)
df_prep.head(10)
```

```python
def merge(df):
  data = pd.read_csv('Insee_Data.csv', sep=';')
  data['code_commune'] = data['Code']
  data = data.drop(columns=['Code', 'Libellé'])
  data['code_commune'] = data['code_commune'].astype(str)
  data.info()
  return pd.merge(df, data, how="left", on='code_commune')
```

```python
df_merge = merge(df_prep)
df_merge.shape
```
