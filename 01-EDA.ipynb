```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
```

## Loading Data

```python
merged = pd.read_csv('df_merged.csv')
```

```python
merged.head(20)
```

```python
merged.isnull().sum().sum()
```

```python
merged["valeur_fonciere"].describe()
```

```python
dis = merged[merged["valeur_fonciere"]< 1200000]
sns.displot(dis, x="valeur_fonciere", hue="type_local")
```

```python
corr = merged.corr()

mask = np.triu(np.ones_like(corr, dtype=bool))
f, ax = plt.subplots(figsize=(11, 9))
cmap = sns.diverging_palette(230, 20, as_cmap=True)

sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
sns.pairplot(merged, hue="type_local", corner=True)
```
